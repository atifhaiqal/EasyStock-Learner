```python
import finnhub
finnhub_client = finnhub.Client(api_key="ctkp081r01qn6d7j5lt0ctkp081r01qn6d7j5ltg")
```

```python
symbols = sp500_financials['Symbol']

# set a new pandas dataframe for the recommendations 
recommendation = pd.DataFrame(symbols)
recommendation[['Consensus', 'Consensus Score']] = None

recommendation
```

```python
def calculate_consensus(ratings):
    weights = {'strongBuy': 5, 'buy': 4, 'hold': 3, 'sell': 2, 'strongSell': 1}
    
    total_weighted_score = 0
    total_ratings = 0
    for key, weight in weights.items():
        total_weighted_score += ratings[key] * weight
        total_ratings += ratings[key]
    
    if total_ratings == 0:
        return None  # Avoid division by zero

    consensus_score = total_weighted_score / total_ratings

    # Classify consensus
    if consensus_score >= 4.5:
        consensus_label = "Strong Buy"
    elif consensus_score >= 3.5:
        consensus_label = "Buy"
    elif consensus_score >= 2.5:
        consensus_label = "Hold"
    elif consensus_score >= 1.5:
        consensus_label = "Sell"
    else:
        consensus_label = "Strong Sell"

    return consensus_score, consensus_label
```

```python
ratings = finnhub_client.recommendation_trends('AAPL')
consensus_score, consensus_label = calculate_consensus(ratings[0])

print(f"Consensus Score: {consensus_score:.2f}")
print(f"Consensus Rating: {consensus_label}")
```

```python
for s in symbols:
    try: 
        ratings = finnhub_client.recommendation_trends(s)

        if ratings:
            consensus_score, consensus_label = calculate_consensus(ratings[0])

            recommendation.loc[recommendation['Symbol'] == s, 'Consensus'] = consensus_label  
            recommendation.loc[recommendation['Symbol'] == s, 'Consensus Score'] = consensus_score
        else:
            print(f"Warning: No ratings available for {s}")
    except ReadTimeout:  # Handle timeout exceptions
        print(f"Timeout error encountered for {s}. Retrying...")
        time.sleep(5)  # Add a small delay before retrying the current symbol
        continue
        
    except finnhub.FinnhubAPIException as e: 
        if e.status_code == 429:
            print(f"API limit reached. Exiting loop.")
            break  # Exit the loop if the API limit is reached
        else:
            print(f"Error: Unable to pull data for {s} - {e}")
            continue  # Continue with the next iteration if it's another error
            
    except Exception as e:  # Handle any other exceptions
        print(f"An unexpected error occurred for {s}: {e}")
        continue  # Continue with the next symbol

    time.sleep(0.5)
```
