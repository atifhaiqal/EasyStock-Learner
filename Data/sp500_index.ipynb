# List of SP500 Companies
Data source: https://www.kaggle.com/datasets/andrewmvd/sp-500-stocks?resource=download

```python
import pandas as pd 
import numpy as np 
```

```python
sp500_df = pd.read_csv("sp500_companies.csv")
sp500_df
```

```python
sp500_df_cleaned = sp500_df[['Symbol', 'Shortname']]
sp500_df_cleaned = sp500_df_cleaned.rename(columns={"Shortname": "ticker_options"})
sp500_df_cleaned.set_index(['Symbol'], inplace=True, append=False, drop=True)
sp500_df_cleaned
```

```python
sp500_json_cleaned = sp500_df_cleaned.to_json("sp500.json")
sp500_json_cleaned
```
