# OIB Data Preprocessing 
Processing the open investment bank dataset to ease the process of data visualization. 
In the final application of the app, this will not be needed as the data used will be from APIs

### Importing data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Open_Investment_Data_2020-2024.csv")
df.head()
```

## Dropping un-wanted columns

```python
df.drop(['Notes', 'ID', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27'], axis=1, inplace=True)
df.head()
```

### Merging all dividens into 1 catergory

```python
df['Action'].describe()
```

```python
df['Action'] = df['Action'].replace(r'(?i)^dividend.*', 'Dividend', regex=True)
```

```python
df['Action'].describe()
```

### Create new column called share/price (GBP)

```python
df['Exchange rate'] = pd.to_numeric(df['Exchange rate'], errors='coerce')

df['Price / Share (GBP)'] = df['Price / share'] / df['Exchange rate']
```
